```python
model = torch.hub.load('pytorch/vision:v0.6.0', 'deeplabv3_resnet101', pretrained=True).eval()
```

```python
# sample execution (requires torchvision)
from PIL import Image
from torchvision import transforms

input_image = Image.open('./kutta.png')
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(256),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

input_tensor = preprocess(input_image)

input_batch = input_tensor.unsqueeze(0) # create a mini-batch as expected by the model
# move the input and model to GPU for speed if available
if torch.cuda.is_available():
    input_batch = input_batch.to('cuda')
    model.to('cuda')
# ouput_original_image = output.argmax(0).float()
```

```python
from src.biggan import BigGAN256
import torch 
import torchvision
from scipy.stats import truncnorm
from torch.autograd import Variable
import matplotlib.pyplot as plt
```

```python
truncation = torch.clamp(torch.tensor(0.4), min=0.02+1e-4, max=1.0-1e-4).float()  # truncation is 0.4
c = torch.tensor((260,)).long()  # class_label is 260
z = truncation * torch.as_tensor(truncnorm.rvs(-2.0, 2.0, size=(1, 140))).float()

z.requires_grad = True

biggan = BigGAN256()

biggan.load_state_dict(torch.load("biggan256-release.pt"))
```

```python
biggan.eval()
optimizer = torch.optim.SGD([z], lr=0.1)

# create a color pallette, selecting a color for each class
palette = torch.tensor([2 ** 25 - 1, 2 ** 15 - 1, 2 ** 21 - 1])
colors = torch.as_tensor([i for i in range(21)])[:, None] * palette
colors = (colors % 255).numpy().astype("uint8")
    
plt.subplot(1,2,1)
plt.imshow(input_image)
    
plt.subplot(1,2,2)
r = Image.fromarray(model(input_batch)['out'][0].argmax(0).float().byte().cpu().numpy()).resize(input_image.size)
r.putpalette(colors)
plt.imshow(r)
    
plt.show()

for i in range(10):
#     z.grad = None
    with torch.enable_grad():
        optimizer.zero_grad()
        img = biggan(z, c, truncation.item())
        output_predicted_image = model(0.5 * (img +1))['out'][0]
        output_original_image = model(input_batch)['out'][0]
#     img = Variable(0.5 * (img +1), requires_grad=True)
#     output_predicted_image = output.argmax(0).float()
        loss = torch.nn.MSELoss()
        cost = loss(output_predicted_image, output_original_image)
        print("cost",i," is ", cost)
    
    plt.subplot(1,2,1)
    pil = torchvision.transforms.ToPILImage()((0.5 * (img.data + 1)).squeeze())
    plt.imshow(pil)
    
    plt.subplot(1,2,2)
    r = Image.fromarray(output_predicted_image.argmax(0).float().byte().cpu().numpy())
    r.putpalette(colors)
    plt.imshow(r)
    
    plt.show()
    
    cost.backward()
    optimizer.step()
```
